```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame(data=pd.read_csv('cars.ORG.csv'))
df.head(-3)
```

```python
df_new=df[['Dimensions.Height','Dimensions.Length','Dimensions.Width','Engine Information.Number of Forward Gears',
          'Fuel Information.City mpg','Engine Information.Engine Statistics.Horsepower','Identification.Make']]
```

```python
df_new
```

```python
x=df_new.drop('Identification.Make',axis=1)
x
```

```python
y=df_new['Identification.Make']
y
```

```python
df_new.columns
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=100)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3)
```

```python
model.fit(xtrain,ytrain)
```

```python
print(model.feature_importances_)
```

```python
feature_importances=pd.Series(model.feature_importances_,index=df_new.columns)\
                      .sort_values(ascending=False)
```
