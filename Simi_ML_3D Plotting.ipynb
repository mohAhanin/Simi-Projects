```python
import pandas as pd
from sklearn import preprocessing 
from sklearn import decomposition
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('cars.ORG.csv')
```

```python
dataframe=pd.DataFrame({
    'Dimensions.Height':dataset.iloc[:1000,0].values,
    'Dimensions.Length':dataset.iloc[:1000,1].values,
    'Dimensions.Width':dataset.iloc[:1000,2].values,
    'City mpg':dataset.iloc[:1000,8].values,
    'Engine Information.Engine Statistics.Torque':dataset.iloc[:1000,17].values,
    'class':dataset.iloc[:1000,13].values,
    })
```

```python
dataframe.head(-3)
```

```python
x=dataframe.iloc[:,0:5].values
y=dataframe.iloc[:,5].values
pca=decomposition.PCA(n_components=3)
pca.fit(x)
```

```python
x=pca.transform(x)
le=preprocessing.LabelEncoder()
y=le.fit_transform(y)
classes=le.classes_
```

```python
colors=['blue','orange','red','pink','purple','green',\
'brown','gray','olive','cyan','tan','gold','coral','tomato'\
,'salmon','darkgreen','yellow','rosybrown','peru','khaki']
```

```python
fig=plt.figure(1,figsize=(10,10))
plt.clf()
ax=Axes3D(fig)
fig.add_axes(ax)
plt.cla()
for i,color in enumerate(colors):
    ax.scatter(x[y==i,0],x[y==i,1],x[y==i,2],c=color)
plt.show()
```
